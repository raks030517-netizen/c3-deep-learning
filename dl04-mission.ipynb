```python
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.datasets import fashion_mnist, cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
```

```python
# =========================================================
# 교재 08-2 CNN 재현
# =========================================================

tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

# reshape 4D 변환
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

model = Sequential([
    Input(shape=(28, 28, 1)),

    # Conv + Pool × 2
    Conv2D(32, kernel_size=3, activation='relu', padding='same'),
    MaxPooling2D(pool_size=2),

    Conv2D(64, kernel_size=3, activation='relu', padding='same'),
    MaxPooling2D(pool_size=2),

    # Dense + Dropout + EarlyStopping
    Flatten(),
    Dense(100, activation='relu'),
    Dropout(0.4),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer=Adam(),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stopping_cb = EarlyStopping(
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=64,
    callbacks=[early_stopping_cb]
)

# model.summary() 스크린샷
model.summary()

val_loss, val_accuracy = model.evaluate(X_test, y_test)

print("val_accuracy:", val_accuracy)

# val_accuracy ≥ 0.91
print("val_accuracy >= 0.91:", val_accuracy >= 0.91)
```

```python
# =========================================================
# 첫 번째 Conv 필터 수 실험
# =========================================================

tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)


def make_cnn(first_filter):
    model = Sequential([
        Input(shape=(28, 28, 1)),

        # Conv2D(32) → Conv2D(64) 교체
        Conv2D(first_filter, kernel_size=3, activation='relu', padding='same'),
        MaxPooling2D(pool_size=2),

        Conv2D(64, kernel_size=3, activation='relu', padding='same'),
        MaxPooling2D(pool_size=2),

        Flatten(),
        Dense(100, activation='relu'),
        Dropout(0.4),
        Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer=Adam(),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model


filter_results = []

for first_filter in [32, 64]:
    tf.keras.utils.set_random_seed(42)

    model = make_cnn(first_filter)

    early_stopping_cb = EarlyStopping(
        patience=3,
        restore_best_weights=True
    )

    history = model.fit(
        X_train,
        y_train,
        validation_data=(X_test, y_test),
        epochs=20,
        batch_size=64,
        callbacks=[early_stopping_cb],
        verbose=0
    )

    val_loss, val_accuracy = model.evaluate(X_test, y_test, verbose=0)

    filter_results.append({
        "첫 번째 Conv 필터 수": first_filter,
        "총 파라미터 수": model.count_params(),
        "val_accuracy": val_accuracy,
        "stopped_epoch": early_stopping_cb.stopped_epoch
    })


# 32 vs 64 필터 비교 결과표
# val_accuracy, 총 파라미터 수 포함
df_filter = pd.DataFrame(filter_results)
display(df_filter)


# 파라미터 수 변화 계산
param_32 = df_filter.loc[
    df_filter["첫 번째 Conv 필터 수"] == 32,
    "총 파라미터 수"
].values[0]

param_64 = df_filter.loc[
    df_filter["첫 번째 Conv 필터 수"] == 64,
    "총 파라미터 수"
].values[0]

param_diff = param_64 - param_32

print("32필터 총 파라미터 수:", param_32)
print("64필터 총 파라미터 수:", param_64)
print("파라미터 수 증가량:", param_diff)


# 성능 비교
acc_32 = df_filter.loc[
    df_filter["첫 번째 Conv 필터 수"] == 32,
    "val_accuracy"
].values[0]

acc_64 = df_filter.loc[
    df_filter["첫 번째 Conv 필터 수"] == 64,
    "val_accuracy"
].values[0]

acc_diff = acc_64 - acc_32

print("32필터 val_accuracy:", acc_32)
print("64필터 val_accuracy:", acc_64)
print("val_accuracy 차이:", acc_diff)


# 실험 관찰 한 문장
print(
    "관찰: 첫 번째 Conv 필터 수를 32에서 64로 늘리면 총 파라미터 수는 증가하지만, "
    "정확도 향상은 실험 결과에 따라 제한적일 수 있다."
)
```

```python
# =========================================================
# CIFAR-10 CNN 확장
# =========================================================

tf.keras.utils.set_random_seed(42)

(X_train, y_train), (X_test, y_test) = cifar10.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

y_train = y_train.reshape(-1)
y_test = y_test.reshape(-1)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

model = Sequential([
    # 3채널 입력
    Input(shape=(32, 32, 3)),

    # 동일 구조
    Conv2D(32, kernel_size=3, activation='relu', padding='same'),
    MaxPooling2D(pool_size=2),

    Conv2D(64, kernel_size=3, activation='relu', padding='same'),
    MaxPooling2D(pool_size=2),

    Flatten(),
    Dense(100, activation='relu'),
    Dropout(0.4),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer=Adam(),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stopping_cb = EarlyStopping(
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=64,
    callbacks=[early_stopping_cb]
)

model.summary()

test_loss, test_accuracy = model.evaluate(X_test, y_test)

# CIFAR-10 test_accuracy 출력
print("test_accuracy:", test_accuracy)

# test_accuracy > 0.65
print("test_accuracy > 0.65:", test_accuracy > 0.65)

print("""
성능 차이 분석:
Fashion MNIST는 흑백 의류 이미지라 특징 추출이 비교적 쉽다.
반면 CIFAR-10은 컬러 이미지이며 물체 종류와 배경이 다양하다.
따라서 같은 CNN 구조에서도 CIFAR-10의 정확도가 더 낮게 나타난다.
""")

# "Fashion MNIST보다 어려운 이유" 한 문장 기록
print("이유: CIFAR-10은 Fashion MNIST보다 컬러 이미지이고 배경, 물체 형태, 촬영 각도가 복잡해서 같은 CNN 구조를 사용해도 분류가 더 어렵다.")
```
